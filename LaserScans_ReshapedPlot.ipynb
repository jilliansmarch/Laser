**Reshaped Plots**

The .npy files are stored as a 1D array, with time values as the first half and voltage values as the second half. We *must* reshape.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
directory = "/path/to/your/files/"
```

Plot Point (0,0)

```python
data = np.load('Test_x0_y0.npy')

plt.plot(data)
plt.title('First Point')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
plt.grid(True)
plt.show()
```

Reshape and plot point (0,0)

```python
data2 = data.reshape([2, 3840])
plt.plot(data2[0, :], data2[1, :]) 
plt.title('Point (0,0)')
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
```

Reshape all of the point files

```python
import os

reshaped_data = {}

for filename in os.listdir(directory):
    if filename.endswith(".npy"):
        filepath = os.path.join(directory, filename)
        data = np.load(filepath)
        print(f"Loading {filename} with shape {np.shape(data)}")
        try:
            reshaped_data[filename] = data.reshape([2, 3839])
        except ValueError as e:
            reshaped_data[filename] = data.reshape([2, 3840])
        except ValueError as e:
            print(f"Error reshaping {filename}: {e}")

print("Data loading complete. Use 'reshaped_data' to access reshaped arrays.")
```

Plot multiple (reshaped) points on one plot

```python
files_to_plot = ["Test_x0_y0.npy", "Test_x0_y4.npy", "Test_x0_y8.npy", "Test_x0_y12.npy", "Test_x0_y16.npy"]

plt.figure(figsize=(10, 6))
for file in files_to_plot:
    if file in reshaped_data:
        data2 = reshaped_data[file]
        plt.plot(data2[0, :], data2[1, :], label=file)
    else:
        print(f"File {file} not found in reshaped_data.")

plt.title("Early Points")
plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')
#plt.xlim([1.5e-5, 3.5e-5])
plt.legend() 
plt.show()
```
